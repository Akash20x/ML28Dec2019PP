```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use("dark_background")
```

```python
arr = np.arange(10)
```

```python
arr_sq = arr**2
```

```python
plt.figure(figsize=(4,4))

plt.plot(arr, arr, color='red', marker="o", label="y=x")
plt.plot(arr, arr_sq, color='green', marker="*", label = "y = xsqaure")
plt.xlabel("Range 0-10")
plt.ylabel("Values")
plt.title("Comparison")
plt.legend()
# plt.axis("off")
plt.savefig("figure.png")
plt.show()
```

# Bar Graph

```python
sales = np.array([20,30,10,5, 50])
sales_2 = np.array([40,20,5,30,22])
x = np.arange(2000,2005)
```

```python
plt.bar(x, sales, width=0.2)
plt.bar(x+0.2, sales_2, width =0.2)
```

## Scatter Plot

```python
plt.scatter(arr, arr)
plt.scatter(arr, arr_sq)
```

```python
X = np.random.randn(100, 2)
Y = np.random.randn(100, 2)*2 + 10 #creating a new normal distribution with std = 2 and mean as 10
```

```python
x = X[:, 0]
y = X[:, 1]
```

```python
plt.scatter(x, y, color="yellow")
plt.scatter(Y[:,0], Y[:,1], color="red")
```

## Histogram

```python
plt.hist(Y[:,1], bins=30)
plt.show()
```
