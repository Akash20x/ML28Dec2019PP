# SVM

```python
class SVM():
    def __init__(self, C = 1.0):
        self.W = 0.0
        self.b = 0.0
        self.C = C
        
    def hinge_loss(self, W, b, X, y ):
        loss = 0.0
        
        loss += 0.5 * np.dot(W,W.T)
        
        m = X.shape[0]
        
        for i in range(m):
            ti = y[i]*(np.dot(W, X[i].T) + b )
            loss += self.C * ( max(0, (1- ti)) )
        
        return loss[0][0]
    
    
    def fit(self, X, y, lr = 0.00001, max_itr = 700):
    
        n = X.shape[1]
        m = X.shape[0]
        
        W = np.zeros((1,n))
        b = 0.0
        
        
        losses = []
        
        for p in range(max_itr):
            
            l = self.hinge_loss(W,b,X,y)
            losses.append(l)
            
            gradw = np.zeros((1,n))
            gradb = 0
            
            gradw += 0.5*W
            
            for i in range(m):
                ti = y[i]*(np.dot(W, X[i].T) + b )
                if (ti>=1):
                    gradw +=0
                    gradb +=0
                else:
                    gradw -= self.C*y[i]*X[i]
                    gradb -= self.C*y[i]
        
            W = W - lr*gradw
            b = b - lr*gradb
        self.W = W
        self.b = b
        
        return W, b, losses
```

# Convert Data for One-Vs-One Classifier

```python
M = X.shape[0]
CLASSES = len(np.unique(y))
```

```python
def classWiseData(X, y):
    data = {}
    
    for i in range(CLASSES):
        data[i] = []
        
    for i in range(X.shape[0]):
        data[y[i]].append(X[i])
        
    for k in data:
        data[k] = np.array(data[k])
        
    return data
```

```python
data = classWiseData(X, y)
```

## Get PairWise Data

```python
def getPairWiseForSVM(d1, d2):
    l1, l2 = d1.shape[0], d2.shape[0]
    samples = l1+l2
    features = d1.shape[1]
    
    data_pair = np.zeros((samples, features))
    data_labels = np.zeros((samples,))
    
    data_pair[:l1, :]  = d1
    data_pair[l1:, :] = d2
    
    data_labels[:l1] = -1
    data_labels[l1:] = 1
    
    return data_pair , data_labels
```

```python
d, l = getPairWiseForSVM(data[0], data[1])
```

# Train Classifiers

```python
mySVM = SVM()
w,b,losses = mySVM.fit(d,l)
```

```python
plt.plot(losses)
```

# Train NC2 Classifiers

```python
def trainSVM(X, y ):
    svm_classifiers = {}
    
    for i in range(CLASSES):
        svm_classifiers[i] = {}
        
        for j in range(i+1, CLASSES):
            xpair, ypair = getPairWiseForSVM(data[i], data[j])
            w, b, l = mySVM.fit(xpair, ypair)

            svm_classifiers[i][j] = [w,b]
            
            plt.plot(l)
            plt.show()

    return svm_classifiers
```

```python
svm_classifiers = trainSVM(X, y)
```

# Prediction

```python
def binaryPredict(x, w, b):
    
    x = np.reshape(x, (1,-1))
    w = np.reshape(w, (1,-1))
    
    
    z = np.dot(w,x.T) +b
    if z>=0:
        return 1
    else:
        return -1
```

```python
def predict(x):
    
    count = np.zeros((CLASSES,))
    
    for i in range(CLASSES):
        for j in range(i+1, CLASSES):
            w, b =svm_classifiers[i][j]
            z = binaryPredict(x,w,b)
            
            if z==1:
                count[j] +=1
            else:
                count[i] +=1
    final_pred = np.argmax(count)
    return final_pred
```

```python
predict(X[111])
```

```python
plt.imshow(X[111].reshape((32,32,3)))
```

```python
def accuracy(X, y ):
    y_pred = []
    for i in range(X.shape[0]):
        p = predict(X[i])
        y_pred.append(int(p))
        
    y_pred = np.array(y_pred)
    acc = (y_pred == y).sum()/y.shape[0]
    return acc
```

```python
accuracy(X, y)
```
