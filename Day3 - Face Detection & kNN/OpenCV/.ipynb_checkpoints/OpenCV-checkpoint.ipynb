```python
group_photo = plt.imread("new_pic.jpeg")
```

```python
plt.imshow(group_photo)
```

```python
group_photo.shape
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = face_cascade.detectMultiScale(group_photo, 1.1, 5)
```

```python
for face in faces:
    x, y, w, h = face
    group_photo = cv2.rectangle(group_photo, (x, y), (x+w,y+h), (255,255,255), 2)
```

```python
plt.imshow(group_photo)
```

# Face Detection on WebCam

```python
cam = cv2.VideoCapture(0)

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
    

while True:
    ret , frame = cam.read()
    
    if ret == False:
        continue
    
        
    faces = face_cascade.detectMultiScale(frame, 1.5, 5)
    
    for face in faces:
        x, y, w, h = face
        frame = cv2.rectangle(frame, (x, y), (x+w,y+h), (255,255,255), 2)
    
    
    cv2.imshow("Face Detection", frame)
    
    key_pressed = cv2.waitKey(1) & 0xFF
    if key_pressed == ord('q'):
        break
    
cam.release()
cv2.destroyAllWindows()
```
