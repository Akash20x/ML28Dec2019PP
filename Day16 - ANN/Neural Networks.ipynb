```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = Sequential()
```

```python
model.add( Dense(units= 16 , activation="relu", input_shape= (2,), name="first" )  )
model.add( Dense(units= 4 , activation="relu", name="second") )
model.add( Dense(units= 1 , activation="sigmoid", name="output"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(500, 2, centers=2, random_state=5 )
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
plt.scatter(X[:,0], X[:, 1], c=y)
plt.show()
```

```python
hist = model.fit(X_train, y_train, batch_size=32, epochs=100, validation_data= (X_test, y_test))
```

```python
res = hist.history
```

```python
plt.plot(res['val_loss'], label = "Val Loss")
plt.plot(res['loss'], label = "Train loss")
plt.legend()
plt.show()
```

```python
plt.plot(res['val_acc'], label = "Val Acc")
plt.plot(res['acc'], label = "Train Acc")
plt.legend()
plt.show()
```
