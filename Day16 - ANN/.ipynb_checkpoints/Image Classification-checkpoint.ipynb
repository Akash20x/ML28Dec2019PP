```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import random
```

```python
dic = {
    'cats' : 0, 'dogs' :1, 'horses':2, 'humans': 3
}
```

```python
categories = os.listdir("./images")
```

```python
X = []
y = []
for c in categories:
    if "." not in c:
        all_imgs = os.listdir("./images/"+c)
        
        
        for i in all_imgs:
            path = "./images/"+c +"/"+i

            img = plt.imread(path)
            img = cv2.resize(img, (50,50))
            X.append(img)
            y.append(dic[c])
            
```

```python
len(X)
```

```python
X = np.array(X)
```

```python
X.shape
```

```python
X = X.reshape((808, -1))
```

```python
X.shape
```

```python
y = np.array(y)
```

```python
y.shape
```

```python
data  = np.hstack((X,y.reshape((-1,1)))) 
```

```python
data.shape
```

```python
np.random.shuffle(data)
```

```python
X = data[:, :-1]
y = data[:, -1]
```

```python
X.shape
```

```python
X = X/255.
```

```python
y.shape
```

```python
for i in range(10):
    plt.imshow(X[i].reshape((50,50,3)))
    plt.show()
```

```python
X.shape
```

```python
y.shape
```

## Keras Model

```python
from keras.models import Sequential
from keras.layers import Dense

from keras.utils import to_categorical
from keras.models import load_model

from keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
model = Sequential()
model.add(Dense(units = 1000, activation="relu", input_shape=(X.shape[1], )))
model.add(Dense(units = 256 , activation="relu"))
model.add(Dense(units = 64, activation="relu" ))
model.add(Dense(units = 4, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
y_oht = to_categorical(y)
```

```python
y_oht.shape
```

```python
modelcheckpoint = ModelCheckpoint("./best_model.h5", monitor="val_acc", verbose=1, save_best_only=True, )
```

```python
early_stop = EarlyStopping(monitor="val_acc", min_delta=0.005, patience=10, verbose=1, restore_best_weights=True)
```

```python
mycallbacks = [modelcheckpoint, early_stop]
```

```python
hist = model.fit(X, y_oht, batch_size=256, epochs=120, validation_split=0.1, callbacks= mycallbacks )
```

```python
res = hist.history
```

```python
plt.plot(res['val_loss'], label = "Val Loss")
plt.plot(res['loss'], label = "Train loss")
plt.legend()
plt.show()
```

```python
plt.plot(res['val_acc'], label = "Val Acc")
plt.plot(res['acc'], label = "Train Acc")
plt.legend()
plt.show()
```
