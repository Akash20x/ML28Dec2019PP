## What does hyperplane mean ?

<img src="img/hyperplanes.jpg" alt="Hyperplanes" style="width: 250px;"/>

A hyperplane is plane of _n-1_ dimensions in _n_ dimensional feature space, that separates the two classes. 
For a 2-D feature space, it would be a line and for a 3-D Feature space it would be plane and so on.

<img src="img/3d_hyperplane.png" alt="Hyperplanes" style="width: 200px;"/>



A hyperplane is able to separate classes if for all points -

#### **_w_ x** + b > 0 
(For data points in class 1)  
#### **_w_ x** + b < 0 
(For data points in  class 0)

## Maximum Margin Hyperplane 

An optimal hyperplane best separates our data so that the distance/margin from nearest points(called Support Vectors) in space to itself is maximized.

## Generate Dataset

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_classification(n_samples=400, n_features=2, n_redundant=0, n_classes=2,
                    n_clusters_per_class=1, random_state=3)
```

```python
plt.scatter(X[:, 0], X[:,1], c = y)
plt.show()
```

```python
y[y==0] = -1
```

```python
class SVM():
    def __init__(self, C = 1.0):
        self.W = 0.0
        self.b = 0.0
        self.C = C
        
    def hinge_loss(self, W, b, X, y ):
        loss = 0.0
        
        loss += 0.5 * np.dot(W,W.T)
        
        m = X.shape[0]
        
        for i in range(m):
            ti = y[i]*(np.dot(W, X[i].T) + b )
            loss += self.C * ( max(0, (1- ti)) )
        
        return loss[0][0]
    
    
    def fit(self, X, y, lr = 0.001, max_itr = 80):
    
        n = X.shape[1]
        m = X.shape[0]
        
        W = np.zeros((1,n))
        b = 0.0
        
        
        losses = []
        
        for p in range(max_itr):
            
            l = self.hinge_loss(W,b,X,y)
            losses.append(l)
            
            gradw = np.zeros((1,n))
            gradb = 0
            
            gradw += 0.5*W
            
            for i in range(m):
                ti = y[i]*(np.dot(W, X[i].T) + b )
                if (ti>=1):
                    gradw +=0
                    gradb +=0
                else:
                    gradw -= self.C*y[i]*X[i]
                    gradb -= self.C*y[i]
        
            W = W - lr*gradw
            b = b - lr*gradb
        self.W = W
        self.b = b
        
        return W, b, losses
        
```

```python
svm = SVM(C= 1)
```

```python
W,b,losses = svm.fit(X, y )
```

```python
plt.plot(losses)
```

```python
def plotHyperPlane(w1, w2, b, X, y):
    plt.figure(figsize=(12,12))
    x_1 = np.linspace(-3, 4, 30)
    x_2 = -1*(b+w1*x_1)/ w2
    
    x_p = -1*(b+w1*x_1 +1)/ w2
    x_n = -1*(b+w1*x_1 -1)/ w2
    
    plt.plot(x_1,x_2)
    plt.plot(x_1, x_p)
    plt.plot(x_1, x_n)
    plt.scatter(X[:, 0], X[:,1], c = y)
    plt.show()
```

```python
plotHyperPlane(W[0][0], W[0][1],  b, X,y)
```
