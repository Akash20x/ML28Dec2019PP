```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
df_x = pd.read_csv("poly_dataset/weightedX.csv").values
df_y = pd.read_csv("poly_dataset/weightedY.csv").values
```

```python
plt.scatter(df_x, df_y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(df_x, df_y)
```

```python
lr.score(df_x, df_y)
```

```python
y = lr.predict(df_x)
```

```python
plt.plot(df_x, y, color='red')
plt.scatter(df_x, df_y)
plt.show()
```

```python
x_square = df_x**2
```

```python
X = np.hstack((df_x, x_square))
```

```python
lr_new = LinearRegression()
lr_new.fit(X, df_y)
```

```python
plt.scatter(df_x, lr_new.predict(X), color="red")
plt.scatter(df_x, df_y)
plt.show()
```

```python
lr_new.score(X, df_y)
```

```python
x_cube = df_x**3
```

```python
X = np.hstack((X,x_cube ))
```

```python
lr_cube = LinearRegression()
lr_cube.fit(X, df_y)
```

```python
plt.scatter(df_x, lr_cube.predict(X), color="red")
plt.scatter(df_x, df_y)
plt.show()
```

```python
lr_cube.score(X, df_y)
```
